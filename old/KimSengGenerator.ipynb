```python
#https://stackoverflow.com/questions/23482668/sorting-by-a-custom-list-in-pandas/27255567
# Create the dictionary that defines the order for sorting
sorterIndex = dict(zip(sorter, range(len(sorter))))

# Generate a rank column that will be used to sort
# the dataframe numerically
df['Point_Rank'] = df['KimSeng Food Pick Up Point'].map(sorterIndex, sorter)

# Here is the result asked with the lexicographic sort
# Result may be hard to analyze, so a second sorting is
# proposed next
df.sort_values('Point_Rank', ascending = True, inplace = True)
df.drop('Point_Rank', 1, inplace = True)
```

```python
# change Menu and Customization into list

dishes = df[df.columns[6]].tolist()
customizes = df[df.columns[7]].fillna(' ').tolist()

#write text
today_ = date.today().strftime("%d.%m.%Y")
txt_name = f"{today_} orders.txt"
file_path = os.path.join('output', txt_name)

if not os.path.exists('output'):
	os.makedirs('output')

with open(file_path, 'w', encoding="utf-8") as writer:
	for index, (d, c) in enumerate(zip(dishes, customizes)):
		writer.write(f'{index+1}. \n')
		for dish in d.split(','):
			dish = dish.strip()
			writer.write(dish)
			writer.write('\n')

		for customize in c.split(','):
			customize = customize.strip()
			writer.write(customize + '\n')

		writer.write('\n')
```

```python
#Export as excel 
etoday_ = date.today().strftime("%d.%m.%Y")
excel_name = f"{etoday_} orders.xlsx"
efile_path = os.path.join('output_excel', excel_name)

if not os.path.exists('output_excel'):
	os.makedirs('output_excel')

writer = pd.ExcelWriter(efile_path)
df_style.to_excel(writer, sheet_name='Sheet1', index = False)


#https://towardsdatascience.com/how-to-auto-adjust-the-width-of-excel-columns-with-pandas-excelwriter-60cee36e175e
#format for the column width, .set_column(index_start, index_end, width)
writer.sheets['Sheet1'].set_column(3, 3, 10)
writer.sheets['Sheet1'].set_column(4, 4, 30)
writer.sheets['Sheet1'].set_column(5, 5, 40)

writer.save()
writer.close()
```
