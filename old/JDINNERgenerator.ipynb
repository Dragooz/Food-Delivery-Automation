```python
#read the file name in the input txt file
# order_name = 'Kim Seng KK6 外送 31_3_2022 (Responses) - Form Responses 1.csv'
input_path = os.path.join('input',"filename.txt")

f = open(input_path,"r", encoding='utf-8')
order_name = f.readlines()[0] 

df = pd.read_csv(os.path.join('input', order_name)  +'.csv')
f.close()
```

```python
#https://stackoverflow.com/questions/23482668/sorting-by-a-custom-list-in-pandas/27255567
# Create the dictionary that defines the order for sorting
sorterIndex = dict(zip(sorter, range(len(sorter))))

# Generate a rank column that will be used to sort
# the dataframe numerically
df['Point_Rank'] = df['DINNER Food Pick Up Point'].map(sorterIndex, sorter)

# Here is the result asked with the lexicographic sort
# Result may be hard to analyze, so a second sorting is
# proposed next
df.sort_values(['Point_Rank','Menu (Select One)'], 
               ascending = [True, True], inplace = True)
df.drop('Point_Rank', 1, inplace = True)
#insert a column of packet numbering
df.insert(3, "No.", range(1,len(df['Email Address'])+1,1))
#create a list of all unique locations.
unique_points = df['DINNER Food Pick Up Point'].unique()

#Generate a column for location name conversion
df['Location'] = df['DINNER Food Pick Up Point'].map(kk_dict, sorter)
#Generate a column for Flavour name conversion
df['Flavour'] = df['Menu (Select One)'].map(flavour_dict, sorter)

#create a list of all unique locations.
unique_locations = df['Location'].unique()
#create a list of all unique locations.
unique_flavours = df['Flavour'].unique()
```
